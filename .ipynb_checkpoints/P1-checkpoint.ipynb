# Proceso de Wiener

Vamos a buscar como es el espectro de un proceso de Wiener unidimensional.

$
\dot{x}(t) = \sqrt{\epsilon} \ \xi(t)
$

Numéricamente lo podemos resolver:

$
x(t+\Delta) - x(t) = \sqrt{\epsilon}\sqrt{\Delta} \ \eta(t)
$

donde $\eta$ es una v.a. normal:

$
<\eta> = 0 \\
<\eta^2> = 1
$

Para el tiempo $t = n\Delta$ y tomando x(0) = 0 podemos escribir:

$$
x(t = n\Delta) = \sum_{i=0}^n \sqrt{\epsilon}\sqrt{\Delta} \ \eta_i(t)
$$

Esta forma es util para resolverlo a la forma "python". Vamos a generar primero todos los número $\eta_i$ de todas las realizaciones y después hacemos todas las sumas cumulativas.

Vamos a hacer N realizaciones de NT pasos hasta el tiempo T.

```python
from numpy.random import normal

def realizaciones(e,dt,N,NT):
    e  = sqrt(e)
    dt = sqrt(dt)
    # Generamos todos los numeros aleatorios de todas las realizaciones y los multiplicamos por las ctes.
    # N*NT numeros aleatorios Gaussianos. Luego los convertimos una matriz NxNT. Cada fila es una realizacion
    I = (normal(0,1,NT*N)*(e*dt)).reshape(N,NT)  
    # Hacemos las sumas cumulativas en cada fila. Estas son cada realizacion.  
    X = cumsum(I,axis=1)  # axis=1 indica que las sumas se hacen por fila
    return(X)

N  = 100     # Numero de realizaciones
NT = 10000   # Numero de iteraciones por realizacion
T  = 10000   # Tiempo
dt = T/NT    # paso de tiempo
e  = .1      # epsilon

X = realizaciones(e,dt,N,NT)    # Calulamos todas las realizaciones
```

Vemos cómo quedan las primeras 10 realizaciones

```python
time = arange(0,T,dt)  # Vector de tiempos
figure(figsize=(9,5))
title('10 realizaciones', fontsize=18)
for i in range(10):
    plot(time,X[i])

xlabel('$t$', fontsize=18)
ylabel('$x(t)$', fontsize=18)
tick_params(labelsize=14)
```

## Espectro

El espectro lo calculamos:

$$
S(\omega) = \lim_{T\to\infty} S(\omega,T) = \lim_{T\to\infty} \left<\frac{1}{T}\left|x(\omega,T)\right|^2\right>
$$

El límite lo hacemos tomando un T grande.

Necesitamos primero calcular la transformada de fourier de cada realización.

$
x(\omega,T) = \mathfrak{F} [x(t,T)]
$

Vamos a usar las librerias de scipy.fftpack. 

Luego calculamos los cuadrados de las transformadas $\left|x(\omega,T)\right|^2$ y hacemos los promedios para cada $\omega$


```python
from scipy.fftpack import fft, fftfreq

def espectro(X, NT, dt, T):
    Xf = abs(fft(X,axis=1))   # Calculamos la fft de cada realizacion y nos quedamos con su valor absoluto
    freq = fftfreq(NT, d=dt)  # Calculamos las frecuencias
    Xf2 = Xf**2               # Elevamos al cuadrado de cada transformada

    S = mean(Xf2/T, axis=0)   # axis=0 indica que el promedio se hace por columnas, o sea, para cada frecuencia
    
    return S, Xf2, freq

S, Xf2, freq = espectro(X, NT, dt, T)
```

Vemos cómo quedan las primeras 10 trasformadas y el promedio.

Graficamos sólo para frecuencias positivas porque es simétrico respecto al origen

```python
figure(figsize=(9,5))
title('Espectro - N={}'.format(N), fontsize=18)
for i in range(10):
    if i == 0: lab = '$x^2(\omega,T)/T$'
    else: lab=None
    plot(freq[:NT//2],Xf2[i][:NT//2]/T, lw=1, alpha=.3,label=lab)
plot(freq[:NT//2],S[:NT//2], lw=3, c='r', label='$S(\omega)$')

yscale('log')
xlabel("$\omega$",fontsize=16)
ylabel("$S(\omega)$",fontsize=16)
tick_params(labelsize=14)
legend(fontsize=16)
```

Para el proceso de Wiener vimos que el espectro es proporcional a $1/\omega^2$

$$
S(\omega) \sim \frac{1}{\omega^2}
$$

Veamos si nuestas simulaciones lo cumplen. Conviene calcular el $log$ y hacer un fiteo lineal.  

$$
\log(S) \sim - 2 \ \log(\omega)
$$

Vemos que para altas frecuencias hay una cola donde deja de cumplirse la realción. Vamos a fitear el $\log(S)$ descartando las altas frecuencias (a ojo).

```python
def fiteo(S, freq):
    # nos quedamos solo con las frecuencias positivas y calculamos los logaritmos
    lS = log(S[1:NT//2])
    lf = log(freq[1:NT//2])
    
    M = len(lf)//10                  # el limite donde descartamos la cola
    z = polyfit(lf[:M], lS[:M], 1)   # fiteo lineal de log(s) sin la cola
    
    return lS, lf, z, M

def plotFiteo(lS, lf, z, M, N):
    # Armamos 1/w^2 para graficar
    w = array([0.001,10])  # w
    w2 = 1/w**2           # 1/w2
    
    figure(figsize=(9,5))
    title('{:.1e} realizaciones'.format(N), fontsize=16)
    plot(lf, lS, lw=1, alpha=.5, label='$log(S)$ con altas frec')
    plot(lf[:M], lS[:M], label='$log(S)$ sin altas frec')
    plot(log(w),log(w2), label='$1/\omega^2$')
    plot(lf, z[1]+z[0]*lf, '--', lw=1, label='fiteo m={:.3f}'.format(z[0]))

    xlabel('$\log(\omega)$', fontsize=18)
    ylabel('$\log(S)$', fontsize=18)
    tick_params(labelsize=14)
    grid()
    legend(fontsize=16)

lS, lf, z, M = fiteo(S, freq)
plotFiteo(lS, lf, z, M, N)
```

## Dependencia con el número de realizaciones

Ahora veamos como se comporta el fiteo con el numero de realizaciones. Vamos ver como evoluciona el parámetro del ajuste $m(N)$ variariando N entre 10 y 10^5

$
\log(S) \sim m \ \log(\omega)
$

Veamos primero que pasa con el grafico anterior si tomamos pocas o muchas realizaciones

```python
NT = 1000    # Numero de iteraciones por realizacion
T  = 10000   # Tiempo
dt = T/NT    # paso de tiempo
e  = .1      # epsilon

N = 10
X = realizaciones(e,dt,N,NT)
S, Xf2, freq = espectro(X, NT, dt, T)
lS, lf, z, M = fiteo(S, freq)
plotFiteo(lS, lf, z, M, N)

N = 10**5
X = realizaciones(e,dt,N,NT)
S, Xf2, freq = espectro(X, NT, dt, T)
lS, lf, z, M = fiteo(S, freq)
plotFiteo(lS, lf, z, M, N)
```

Para $N=10^5$ el ruido de $S$ desaparece y podemos hacer los cálculos hasta ahí. 

```python
NT = 1000    # Numero de iteraciones por realizacion
T  = 10000   # Tiempo
dt = T/NT    # paso de tiempo
e  = .1      # epsilon

NN = logspace(2, 5, num=50, dtype=int) # Los numeros de realizaciones que vamos a calcular distribuidos exponencialmente

Z = []   
for N in NN:
    X = realizaciones(e,dt,N,NT)
    S, Xf2, freq = espectro(X, NT, dt, T)
    lS, lf, z, M = fiteo(S, freq)
    Z.append(z[0])   # Vamos guardando las pendientes del fiteo
    
```

```python
figure(figsize=(9,5))
title('Dependencia con numero de realizaciones', fontsize=16)
plot(NN,Z, 'o')
axhline(-2, ls='--', c='r')

xlabel('$N$', fontsize=16)
ylabel('$m(N)$', fontsize=16)
xscale('log')
tick_params(labelsize=14)
grid()
ylim(-2.02,-1.98);
```

Vemos que el exponente tiende casi a -2 como esperabamos. El valor absluto está siempre por debajo de 2, esto probablemente sea porque todavía hay influencia de las altas frecuencias y el corte lo deberíamos haber hecho más grande.
